```python
import numpy as np
import matplotlib.pyplot as plt

# Data for Li-ion
li_data = np.array([
    [0.236, 2.30, 0.694, 0.0725, 129, 113, 49.5],  # Energy (MJ)
    [0.200, 0.385, 0.030, 0.00217, 8.75, 8.59, 4.03],  # Global Warming (kg CO2 eq)
    [0.000489, 0.000707, 0.000109, 7.25E-06, 0.0886, 0.0201, 0.0196],  # Acidification (kg SO2 eq)
    [0.000489, 0.000236, 5.43E-05, 0.00, 0.0375, 0.0859, 0.0141]  # Eutrophication (kg PO4 eq)
])

# Data for Al-ion
al_data = np.array([
    [1.48, 0.550, 0.116, 0.0651, 129, 113, 56.7],  # Energy (MJ)
    [0.0909, 0.0717, 0.0102, 0.00115, 8.75, 8.59, 4.62],  # Global Warming (kg CO2 eq)
    [0.000236, 0.000637, 5.36E-05, 6.45E-06, 0.0886, 0.0201, 0.0251],  # Acidification (kg SO2 eq)
    [0.000236, 0.000106, 2.26E-05, 0.00, 0.0375, 0.0859, 0.0132]  # Eutrophication (kg PO4 eq)
])

# Calculate cumulative sums
Cell_Li = 30 * li_data[:, :4].sum(axis=1)
BMS_Li = li_data[:, 4:6].sum(axis=1)
Manu_Li = li_data[:, 6]

Cell_Al = 40 * al_data[:, :4].sum(axis=1)
BMS_Al = al_data[:, 4:6].sum(axis=1)
Manu_Al = al_data[:, 6]

# Generate cumulative data matrices
Li_cum = np.vstack([
    Cell_Li,
    Cell_Li + BMS_Li,
    Cell_Li + BMS_Li + Manu_Li
]).T

Al_cum = np.vstack([
    Cell_Al,
    Cell_Al + BMS_Al,
    Cell_Al + BMS_Al + Manu_Al
]).T

# Plot settings
names = ["Cell Batch Components", "BMS & Cooling System Process", "Cell Manufacturing"]
ylabels = ["Energy (MJ)", "Global Warming (kg CO2 eq)", "Acidification (kg SO2 eq)", "Eutrophication (kg PO4 eq)"]
ylims = [(0, 500), (0, 50), (0, 0.2), (0, 0.2)]

# Plotting
for i in range(4):
    plt.figure()
    plt.plot([1, 2, 3], Li_cum[i], 'o', label='Li-ion')
    plt.plot([1, 2, 3], Al_cum[i], 'o', label='Al-ion')
    plt.xlim(0, 4)
    plt.xticks([1, 2, 3], names, rotation=20)
    plt.ylim(ylims[i])
    plt.xlabel("Cumulative Summation of Parts")
    plt.ylabel(ylabels[i])
    plt.legend()
    plt.show()

# Calculate percentage differences
Al_norm = Al_cum - Li_cum
perc = Al_norm / Li_cum * 100 + 100

# Plotting percentage differences
# Continue plotting percentage differences
plt.figure(figsize=(10, 5))
plt.plot([1, 2, 3], [100, 100, 100], 'o-', label='Li-ion Baseline')
labels = ["Al-ion Energy", "Al-ion Global Warning CO2", "Al-ion Acidification SO2", "Al-ion Eutrophication PO4"]
for i in range(4):
    plt.plot([1, 2, 3], perc[i], 'o-', label=labels[i])
plt.xlim(0, 4)
plt.xticks([1, 2, 3], names)
plt.xticks(rotation=20)
# plt.xlabel("Cumulative Emissions")
plt.xlabel("")
plt.ylabel("Percentage Relative to Baseline (%)")
plt.ylim(20, 120)
plt.legend(loc='lower right')
plt.title("Percentage Relative to Baseline for Al-ion")
plt.grid(True)
plt.show()
```
