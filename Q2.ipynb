```python
#Variables
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#Q2 i - Li
cells_per_pack_Li = 30
cells_per_pack_Al = 40
total_packs_required_Li = 10000
total_packs_required_Al = 10000
# total_packs_required_Li = 2500
# total_packs_required_Al = 2500
operational_hours_per_day = 20
operational_days_per_week = 7
picking_time_per_cell = 9  # seconds
bin_replacement_time = 15 # Seconds
defect_rate_Li = 1 / 4000
defect_rate_Al = 1 / 2000
n_lines = 1
n_conveyer = 3
n_robot = 1
number_of_components_per_line = n_conveyer+n_robot
probability_of_failure = 0.01
probability_of_failure_per_pack = number_of_components_per_line*probability_of_failure
average_downtime_per_failure = 30 * 60  # converting minutes to seconds
#Cost Analysis
cost_per_conveyer = 10*1e3
cost_per_robot = 125*1e3
conveyer_use_per_hour = 10 
robot_use_per_hour = 10
conveyer_repair_per_hour = 1000
robot_repair_per_hour = 1000
current_cost = 155*1e3
```

```python
# Effective cells needed considering defects
effective_cells_per_pack_Li = cells_per_pack_Li / (1 - defect_rate_Li)
print(f'The total number of packs required including the rate of defect to meet the annual demand is {effective_cells_per_pack_Li}')
# Total effective cells needed
total_effective_cells_Li = effective_cells_per_pack_Li * total_packs_required_Li
print(f'If this value of {effective_cells_per_pack_Li} is converted in the number of cell that needs to manufactured, this is {total_effective_cells_Li}')
print("----------")

# Actual Time
#Total Time per pack:
time_per_pack_Li = effective_cells_per_pack_Li*picking_time_per_cell
print(f'Time per pack is {time_per_pack_Li} s')
#Downtime per pack
downtime_per_second_Li = probability_of_failure*average_downtime_per_failure
print(f'Every hour the machine is going to fail {downtime_per_second_Li}')
time_per_pack_with_bin_Li = time_per_pack_Li+downtime_per_second_Li*number_of_components_per_line+bin_replacement_time
print(f'Thus the real time per pack is {time_per_pack_with_bin_Li}')
operational_time_for_packs_Li = total_packs_required_Li*time_per_pack_with_bin_Li
operational_time_hour_Li = operational_time_for_packs_Li/3600
print(f'Thus the total actual operational hour is {operational_time_hour_Li} hours')
production_rate_actual_hour_Li = total_packs_required_Li/operational_time_hour_Li
print(f'The actual production per hour is {production_rate_actual_hour_Li} unit per hour')
print("----------")
# Number of failures expected over the production of 10,000 packs
expected_failures_Li = total_packs_required_Li * probability_of_failure_per_pack
print(f'As the failure is {probability_of_failure_per_pack}, the toal number of packs that are going to fail are {expected_failures_Li}')
# Total downtime due to failures in hours
total_downtime_hours_Li = ((expected_failures_Li * average_downtime_per_failure)+(expected_failures_Li*bin_replacement_time)) / 3600
# total_downtime_hours_Li = ((expected_failures_Li * average_downtime_per_failure)) / 3600
print(f'Thus, as the downtime per falure is {average_downtime_per_failure/60} minutes, the production line will be stopped for {total_downtime_hours_Li} hours')
print("----------")
# Ideal Time
time_per_pack_ideal_Li = cells_per_pack_Li*picking_time_per_cell+bin_replacement_time
print(f'The ideal time without any losses per pack is {time_per_pack_ideal_Li} seconds')
prodution_rate_ideal_hour_Li = 3600/time_per_pack_ideal_Li
print(f'The ideal production rate is {prodution_rate_ideal_hour_Li} unites per hour')
print("----------")
# Efficiency
Efficiency_Li = production_rate_actual_hour_Li/prodution_rate_ideal_hour_Li
print(f'The efficiency is: {Efficiency_Li*100} % ')
print("----------")
print("Cost Analysis")

#### Aluminium
print('\nALUMINIUM BATTERIES')
# Effective cells needed considering defects
effective_cells_per_pack_Al = cells_per_pack_Al / (1 - defect_rate_Al)
print(f'The total number of packs required including the rate of defect to meet the annual demand is {effective_cells_per_pack_Al}')
# Total effective cells needed
total_effective_cells_Al = effective_cells_per_pack_Al * total_packs_required_Al
print(f'If this value of {effective_cells_per_pack_Al} is converted in the number of cell that needs to manufactured, this is {total_effective_cells_Al}')
print("----------")

# Actual Time
#Total Time per pack:
time_per_pack_Al = effective_cells_per_pack_Al*picking_time_per_cell
print(f'Time per pack is {time_per_pack_Al} s')
#Downtime per pack
downtime_per_second_Al = probability_of_failure*average_downtime_per_failure
print(f'Every hour the machine is going to fail {downtime_per_second_Al}')
time_per_pack_with_bin_Al = time_per_pack_Al+downtime_per_second_Al*number_of_components_per_line+bin_replacement_time
print(f'Thus the real time per pack is {time_per_pack_with_bin_Al}')
operational_time_for_packs_Al = total_packs_required_Al*time_per_pack_with_bin_Al
operational_time_hour_Al = operational_time_for_packs_Al/3600
print(f'Thus the total actual operational hour is {operational_time_hour_Al} hours')
production_rate_actual_hour_Al = total_packs_required_Al/operational_time_hour_Al
print(f'The actual production per hour is {production_rate_actual_hour_Al} unit per hour')
print("----------")
# Number of failures expected over the production of 10,000 packs
expected_failures_Al = total_packs_required_Al * probability_of_failure_per_pack
print(f'As the failure is {probability_of_failure_per_pack}, the toal number of packs that are going to fail are {expected_failures_Al}')
# Total downtime due to failures in hours
total_downtime_hours_Al = ((expected_failures_Al * average_downtime_per_failure)+(expected_failures_Al*bin_replacement_time)) / 3600
# total_downtime_hours_Al = ((expected_failures_Al * average_downtime_per_failure)) / 3600
print(f'Thus, as the downtime per falure is {average_downtime_per_failure/60} minutes, the production line will be stopped for {total_downtime_hours_Al} hours')
print("----------")
# Ideal Time
time_per_pack_ideal_Al = cells_per_pack_Al*picking_time_per_cell+bin_replacement_time
print(f'The ideal time without any losses per pack is {time_per_pack_ideal_Al} seconds')
prodution_rate_ideal_hour_Al = 3600/time_per_pack_ideal_Al
print(f'The ideal production rate is {prodution_rate_ideal_hour_Al} unites per hour')
print("----------")
# Efficiency
Efficiency_Al = production_rate_actual_hour_Al/prodution_rate_ideal_hour_Al
print(f'The efficiency is: {Efficiency_Al*100} % ')
print("----------")
print("\n\nCOST ANALYSIS")

#####COST
# Purchase Cost
Total_cost_conveyer = n_conveyer*cost_per_conveyer
Total_cost_robot = n_robot*cost_per_robot
Total_purchase_cost = ((Total_cost_conveyer+Total_cost_robot)*2*n_lines)-current_cost
print(f'Conveyer cost in pound is {Total_cost_conveyer}, robot is {Total_cost_robot}, thus the total purchase cost is {Total_purchase_cost}')

###### LITHIUM
print(f'\nLITHIUM COST LINE')
# Operational Cost
# Conveyer Operational Cost
conveyer_use_cost_Li = n_conveyer*conveyer_use_per_hour*(operational_time_hour_Li-total_downtime_hours_Li)
print(f'Total Conveyer Cost for use is {conveyer_use_cost_Li} £')
conveyer_repair_cost_Li = n_conveyer*conveyer_repair_per_hour*(total_downtime_hours_Li/number_of_components_per_line)
print(f'Conveyer Repair cost is: {conveyer_repair_cost_Li} £')
#Robot Operational Cost
robot_use_cost_Li = n_robot*robot_use_per_hour*(operational_time_hour_Li-total_downtime_hours_Li)
print(f'Total Robot Cost for use is {robot_use_cost_Li} £')
robot_repair_cost_Li = n_robot*robot_repair_per_hour*(total_downtime_hours_Li/number_of_components_per_line)
print(f'Robot Repair cost is: {robot_repair_cost_Li} £')
total_operational_cost_Li = (conveyer_use_cost_Li + conveyer_repair_cost_Li + robot_use_cost_Li + robot_repair_cost_Li)*n_lines
print(f'Total Operational Cost of all Lithium Line is: {total_operational_cost_Li} £')
total_Li_cost = (Total_purchase_cost+total_operational_cost_Li)
total_Li_cost_hour = (total_operational_cost_Li/operational_time_hour_Li)
print(f'Total Operational cost per hour for Li is: {total_Li_cost_hour}')
print(f'The Total Cost of Li is: {total_Li_cost}')

###### ALUMINIUM
print(f'\nALUMINIUM COST LINE')
# Operational Cost
# Conveyer Operational Cost
conveyer_use_cost_Al = n_conveyer*conveyer_use_per_hour*(operational_time_hour_Al-total_downtime_hours_Al)
print(f'Total Conveyer Cost for use is {conveyer_use_cost_Al} £')
conveyer_repair_cost_Al = n_conveyer*conveyer_repair_per_hour*(total_downtime_hours_Al/number_of_components_per_line)
print(f'Conveyer Repair cost is: {conveyer_repair_cost_Al} £')
#Robot Operational Cost
robot_use_cost_Al = n_robot*robot_use_per_hour*(operational_time_hour_Al-total_downtime_hours_Al)
print(f'Total Robot Cost for use is {robot_use_cost_Al} £')
robot_repair_cost_Al = n_robot*robot_repair_per_hour*(total_downtime_hours_Al/number_of_components_per_line)
print(f'Robot Repair cost is: {robot_repair_cost_Al} £')
total_operational_cost_Al = (conveyer_use_cost_Al + conveyer_repair_cost_Al + robot_use_cost_Al + robot_repair_cost_Al)*n_lines
print(f'Total Operational Cost of all Lithium Line is: {total_operational_cost_Al} £')
total_Al_cost = (Total_purchase_cost+total_operational_cost_Al)
total_Al_cost_hour = (total_operational_cost_Al/operational_time_hour_Al)
print(f'Total Operational cost per hour for Al is: {total_Al_cost_hour}')
print(f'The Total Cost of Al is: {total_Al_cost}')

# # Graph
# Total operational cost
total_operational_cost = total_operational_cost_Li/operational_time_hour_Li + total_operational_cost_Al/operational_time_hour_Al
print(f'\nTotal Operational Cost per hour of every line is: {total_operational_cost} £')

# Calculate cumulative costs over time
time_hours_Li = np.arange(0, operational_time_hour_Li + 1)
time_hours_Al = np.arange(operational_time_hour_Li, operational_time_hour_Al + 1)

# THIS ONE FOR JUST LITHIUM ON ITS OWN
# cost_accumulated_Li = total_operational_cost * time_hours_Li 

cost_accumulated_Li = total_operational_cost * time_hours_Li + Total_purchase_cost
cost_accumulated_Al = cost_accumulated_Li[-1] + total_Al_cost_hour * (time_hours_Al - operational_time_hour_Li)

# THIS ONE FOR JUST LITHIUM ON ITS OWN
# time_hours_combined = time_hours_Li
# cost_accumulated_combined = cost_accumulated_Li

# Combine arrays for plotting
time_hours_combined = np.concatenate([time_hours_Li, time_hours_Al[1:]])
cost_accumulated_combined = np.concatenate([cost_accumulated_Li, cost_accumulated_Al[1:]])

# Plotting the cumulative cost graph
plt.figure(figsize=(10, 5))

# THIS ONE FOR JUST LITHIUM ON ITS OWN
# plt.plot(time_hours_Li, cost_accumulated_Li, label='Cumulative Operational Cost')

plt.plot(time_hours_combined, cost_accumulated_combined, label='Cumulative Operational Cost')
plt.title('Cumulative Operational Cost Over Time')
plt.xlabel('Operational Time (Hours)')
plt.ylabel('Cumulative Cost (£)')
plt.axvline(x=operational_time_hour_Li, color='r', linestyle='--', label='Max Li Operational Hours')
plt.legend()
plt.grid(True)
plt.show()

```
